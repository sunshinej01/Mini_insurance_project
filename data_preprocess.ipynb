```python
#20대에서 50대까지만 추출
import pandas as pd
import numpy as np

# csv파일경로 설정
filename = "2022.11_연령성별담보별 보험회사 계약건수.csv"
df = pd.read_csv(filename)
df = df[df['age_val'] < 60]
df = df[df['age_val'] >= 20]
df.to_csv(filename, index = False)
```

```python
#중분류 설정

df = pd.read_csv(filename)
new_column = []
for value in df['age_val']:
    if 20 <= value < 30:
        new_column.append(np.random.choice(['tour','culture','sports','hobbies'], p=[0.039,0.237,0.535, 0.189 ]))
    elif 30 <= value < 40:
        new_column.append(np.random.choice(['tour','culture','sports','hobbies'], p=[0.086, 0.165, 0.535, 0.214 ]))
    elif 40 <= value < 50:
        new_column.append(np.random.choice(['tour','culture','sports','hobbies'], p=[0.055,0.131, 0.519, 0.295 ]))
    elif 50 <= value:
        new_column.append(np.random.choice(['tour','culture','sports','hobbies'], p=[0.042, 0.098, 0.565, 0.295]))
    else:
        new_column.append('null')  # 범위 밖의 값은 'N'으로 설정
# 새로운 열 추가
df['category'] = new_column
```

```python
#세부분류 설정

#filename = '2022.12_연령성별담보별 보험회사 계약건수.csv'
df = pd.read_csv(filename)

# '세부분류' 값을 지정하는 함수
def assign_subcategory(row):
    if row['category'] == 'hobbies':
        if 20 <= row['age_val'] < 30:
            return np.random.choice(['hiking', 'fishing', 'pet'], p=[0.399, 0.109, 0.492])  # 20세 미만
        elif 30 <= row['age_val'] < 40:
            return np.random.choice(['hiking', 'fishing', 'pet'], p=[0.487, 0.180, 0.333])
        elif 40 <= row['age_val'] < 50:
            return np.random.choice(['hiking', 'fishing', 'pet'], p=[0.523, 0.207, 0.270])
        elif row['age_val'] >= 50:
            return np.random.choice(['hiking', 'fishing', 'pet'], p=[0.613, 0.206, 0.181])
        else:
            return 'null'
    elif row['category'] == 'sports':
        if 20 <= row['age_val'] < 30:
            return np.random.choice(['work_out','yoga','golf'], p=[0.728,0.204,0.068 ])
        elif 30 <= row['age_val'] < 40:
            return np.random.choice(['work_out','yoga','golf'], p=[0.592,0.265,0.143 ])
        elif 40 <= row['age_val'] < 50:
            return np.random.choice(['work_out','yoga','golf'], p=[0.457,0.186,0.357 ])
        elif row['age_val'] >= 50:
            return np.random.choice(['work_out','yoga','golf'], p=[0.435,0.179,0.386])
        else:
            return 'null'
        
    elif row['category'] == 'culture':
        if 20 <= row['age_val'] <30:
            return np.random.choice(['musical','concert','classic'], p=[0.483, 0.353,0.164 ])
        elif 30 <= row['age_val'] < 40:
            return np.random.choice(['musical','concert','classic'], p=[0.449, 0.350, 0.201])
        elif 40 <= row['age_val'] < 50:
            return np.random.choice(['musical','concert','classic'], p=[0.411, 0.353, 0.236 ])
        elif 50 <= row['age_val']:
            return np.random.choice(['musical','concert','classic'], p=[0.349,0.424,0.227])
        else:
            return 'null'
            
    elif row['category'] == 'tour':
        if 20 <= row['age_val'] < 30:
            return np.random.choice(['camping','picnic'], p=[0.654,0.346 ])
        elif 30 <= row['age_val'] < 40:
            return np.random.choice(['camping','picnic'], p=[0.592, 0.408])
        elif 40 <= row['age_val'] < 50:
            return np.random.choice(['camping','picnic'], p=[ 0.594, 0.406])
        elif 50 <= row['age_val']:
            return np.random.choice(['camping','picnic'], p=[0.518, 0.482])
        else:
            return 'null'    
    else:
        return 'null'
# 새로운 '세부분류' 열 추가
df['sub_category'] = df.apply(assign_subcategory, axis=1)

# print(df)
df.to_csv(filename, index = False)
```

```python
# sub_category 기준으로 under sampling

# DataFrame 로드
df = pd.read_csv(filename)

# 각 sub_category의 개수 확인
min_count = df['sub_category'].value_counts().min()

# 언더샘플링
undersampled_df = df.groupby('sub_category').apply(lambda x: x.sample(min_count)).reset_index(drop=True)

# 결과 확인
print(undersampled_df['sub_category'].value_counts())

# 언더샘플링된 DataFrame을 CSV로 저장 (선택 사항)
undersampled_df.to_csv(filename, index=False)
```
